```python
import pandas as pd 
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
data = pd.DataFrame(digits.data)
data
```

```python
data["target"] = digits.target
data
```

```python
x = data.drop("target",axis=1)
x
```

```python
y= data.target
y
```

```python
from sklearn.model_selection import train_test_split
xtrain ,xtest ,ytrain ,ytest =train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
Model_forest = RandomForestClassifier(n_estimators=500)
```

```python
Model_forest.fit(xtrain,ytrain)
```

```python
Model_forest.score(xtrain,ytrain)
```

```python
Model_forest.score(xtest ,ytest )
```

```python
y_prediction = Model_forest.predict(xtest)
y_prediction
```

```python
from sklearn.metrics import confusion_matrix 
confu = confusion_matrix(y_prediction , ytest )
confu
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
sns.heatmap(confu,annot=True)
plt.show()
```
