#### Handling Date

```python
data[["Order Date"]]
```

```python
data["data"]=pd.to_datetime(data["Order Date"],dayfirst=True)
```

```python
data["day"]=data["data"].dt.day
```

```python
data
```

```python
data["month"]=data["data"].dt.month
```

```python
data["year"]=data["data"].dt.year
```

```python
data["moth_year"]=data.data.dt.to_period("M")
```

```python
data
```

```python
yearly_sales=data.groupby("year").Price.sum()
yearly_sales
```

```python
plt.figure(figsize=(10, 6))
yearly_sales.plot(kind="bar", color="lightgreen", edgecolor="black")
plt.xlabel("Year")
plt.ylabel("Total Price")
plt.title("Total Price by Year")
plt.xticks(rotation=45)  
plt.show()
```

```python
day_counts=data.day.value_counts()
day_counts
```

```python
palette = sns.color_palette("husl", len(day_counts))
plt.figure(figsize=(13, 6))
day_counts.plot(kind="bar", color=palette, edgecolor="blue")
plt.xlabel("Day")
plt.ylabel("Count")
plt.title("Day Distribution")
plt.xticks(rotation=45)
plt.show()
```

```python
Month_counts=data.moth_year.value_counts()
Month_counts
```

```python
num_bars = len(Month_counts)
palette = sns.color_palette("flare", num_bars)
plt.figure(figsize=(15, 7))
Month_counts.plot(kind="bar", color=palette, edgecolor="blue")
plt.xlabel("Month-Year")
plt.ylabel("Count")
plt.title("Month Distribution")
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
month_Profit=data.groupby("moth_year").Profit.sum()
month_Profit
```

```python
num_bars = len(month_Profit)
palette = sns.color_palette("mako", num_bars)
plt.figure(figsize=(15, 7))
month_Profit.plot(kind="bar", color=palette, edgecolor="blue")
plt.xlabel("Month-Year")
plt.ylabel("Total Profit")
plt.title("Monthly Profit Distribution")
plt.xticks(rotation=45)

plt.show()
```
